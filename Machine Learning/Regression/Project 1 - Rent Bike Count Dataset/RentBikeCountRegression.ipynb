#Checking data is normally distributed or not

```python
import scipy.stats as stats
plt.figure(figsize=(15,20))
# Iterate over each numerical column and create a separate probplot
for i,f in enumerate(num_c):
  plt.subplot(5,3,i+1)
  stats.probplot(df[f], dist="norm", plot=plt)
  plt.title(f"Probability Plot of {col}") # Add a title for each plot
```

```python
from scipy.stats import shapiro

stat, p_value = shapiro(df[num_c])
print(f'Statistic: {stat}, p-value: {p_value}')

if p_value > 0.05:
    print("The data is normally distributed (fail to reject H0).")
else:
    print("The data is not normally distributed (reject H0).")
```

```python
from scipy.stats import skew, kurtosis

skewness = skew(df[num_c])
kurt = kurtosis(df[num_c])

print(f'Skewness: {skewness}\n\n')
print(f'Kurtosis: {kurt}')
```

```python
X=df.drop('Rented Bike Count',axis=1).values
Y=df['Rented Bike Count'].values
```

```python

print(X[10])
```

```python
print(Y[1])
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[9])],remainder='passthrough')
X=np.array(ct.fit_transform(X))
```

```python
print(X[10])
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
X[:,13]=le.fit_transform(X[:,13])
X[:,14]=le.fit_transform(X[:,14])
```

```python
print(X[10])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,Y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2lr=r2_score(Y_test,y_pred)
r2lr
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=3)
X_poly=poly.fit_transform(X)
lr2=LinearRegression()
lr2.fit(X_poly,Y)
```

```python
y_pred_poly=lr2.predict(poly.fit_transform(X_test))
```

```python
from sklearn.metrics import r2_score
r2pr=r2_score(Y_test,y_pred_poly)
r2pr
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train_t=sc.fit_transform(X_train)
X_test_t=sc.transform(X_test)
```

```python
from sklearn.svm import SVR
svr=SVR(kernel='rbf')
svr.fit(X_train_t,Y_train_t)
```

```python
y_pred_svr=svr.predict(X_test_t)
```

```python
from sklearn.metrics import r2_score
r2svm=r2_score(Y_test,y_pred_svr)
r2svm
```

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
dt.fit(X_train,Y_train)
```

```python
y_pred_dt=dt.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2dt=r2_score(Y_test,y_pred_dt)
r2dt
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(n_estimators=10,random_state=0)
rf.fit(X_train,Y_train)
```

```python
y_pred_randomreg=rf.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2rfr=r2_score(Y_test,y_pred_randomreg)
r2rfr
```

```python
from xgboost import XGBRegressor
xgb=XGBRegressor()
xgb.fit(X_train,Y_train)
```

```python
y_pred_xgb=xgb.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2xgb=r2_score(Y_test,y_pred_xgb)
r2xgb
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gbr=GradientBoostingRegressor(n_estimators=10)
gbr.fit(X_train,Y_train)
```

```python
y_pred_gbr=gbr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2gbr=r2_score(Y_test,y_pred_gbr)
r2gbr
```

```python
model_names=['Linear Regression','Polynomial Regression','Support Vector Regression','Decision Tree Regression','Random Forest Regression','XGBoost Regression','Gradient Boosting Regression']
r2_values=[r2lr,r2pr,r2svm,r2dt,r2rfr,r2xgb,r2gbr]

plt.figure(figsize=(10, 6))
plt.bar(model_names, r2_values, color='indigo')
plt.xlabel('Models')
plt.ylabel('R Score')
plt.title('R Scores of Different Models')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a dot plot
plt.figure(figsize=(10, 6))
plt.plot(model_names, r2_values, 'o', color='blue')
plt.xticks(rotation=45)
plt.xlabel('Models')
plt.ylabel('R Score')
plt.title('R Scores of Different Models')
plt.ylim([0, 1])
plt.grid(True)
plt.show()
```
